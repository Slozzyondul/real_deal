```python
import tabula
import pandas as pd
# Suppress stderr messages

import sys
import warnings
import os


```

```python
warnings.filterwarnings("ignore")

# Redirect stderr to null to suppress warnings
sys.stderr = open(os.devnull, 'w')


pdf_file = "/home/solo/Desktop/2022-1.pdf"

# Read the PDF file and extract text data into a DataFrame
df = tabula.read_pdf(pdf_file, pages="all", output_format="dataframe")


# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(df)

# Reset stderr to its default value
sys.stderr = sys.__stderr__

warnings.resetwarnings()

```
